```python
import pandas as pd

dataset = pd.read_csv('/Users/RodolfoSaldanha/Desktop/DL/2_lecture/Lab/Churn_Prediction_Problem.csv')
dataset.head()
```

```python
dataset.drop(columns=['CustomerId', 'Surname', 'RowNumber'],inplace=True)
dataset.head()
```

```python
cat_columns = ['Geography','Gender']
df_processed = pd.get_dummies(dataset, prefix_sep="_",
                              columns=cat_columns)
```

```python
target = ['Exited ']
```

```python
variables = df_processed[df_processed.columns[~df_processed.columns.isin(target)]]
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data, train_label, test_label = train_test_split(variables,
                                                                  df_processed['Exited '],test_size=0.3)
```

```python
import numpy as np

x_train = np.asarray(train_data)
x_test = np.asarray(test_data)

y_train = np.asarray(train_label).astype('float32')
y_test = np.asarray(test_label).astype('float32')
```

```python
from keras import models
from keras import layers

model = models.Sequential()
model.add(layers.Dense(6, activation='relu', input_dim=13))
model.add(layers.Dense(6, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs=20, batch_size=512)
```
